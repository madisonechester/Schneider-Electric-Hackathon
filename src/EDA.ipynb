```python
from datetime import datetime, timedelta

def get_hourly_array(start_date, end_date):
    
    # Especifica el rango de fechas desde el primer día de 2022 hasta el último
    # start_date = "2022-01-01 00:00:00"
    # end_date = "2022-12-31 23:00:00"
    
    # Crea un rango de fechas cada hora
    start_time_range = pd.date_range(start=start_date, end=end_date, freq='H')
    # array
    start_time_array = start_time_range.strftime('%Y-%m-%d %H:%M:%S%z').to_numpy()

    # entime_range
    start_date_2 = pd.to_datetime(start_date) + timedelta(hours=1) # le sumamos una hora
    start_date_2 = start_date_2.strftime('%Y-%m-%d %H:%M:%S%z')
    end_date_2 = pd.to_datetime(end_date) + timedelta(hours=1)
    end_date_2 = end_date_2.strftime('%Y-%m-%d %H:%M:%S%z')
    end_time_range = pd.date_range(start=start_date_2, end=end_date_2, freq='H')
    end_time_array = end_time_range.strftime('%Y-%m-%d %H:%M:%S%z').to_numpy()
    
    return start_time_array, end_time_array


start_date = df_def_copy['StartTime'].min()
end_date = df_def_copy['StartTime'].max()

print(start_date, type(start_date))

start_time_array, end_time_array = get_hourly_array (start_date, end_date)

print(min(start_time_array), min(end_time_array))

print(max(start_time_array), max(end_time_array))
```

# new way of agregating dat

```python
import pandas as pd
import os

def read_and_concatenate(folder_path):
    # Lists to store DataFrames
    gen_dataframes = []
    load_dataframes = []

    # Iterate over all files in the folder
    for file in os.listdir(folder_path):
        if file.endswith('.csv'):
            file_path = os.path.join(folder_path, file)

            # Read 'gen' files
            if file.startswith('gen'):
                df = pd.read_csv(file_path)
                gen_dataframes.append(df)

            # Read 'load' files
            elif file.startswith('load'):
                df = pd.read_csv(file_path)
                load_dataframes.append(df)

    # Concatenate DataFrames vertically
    gen_concatenated = pd.concat(gen_dataframes, axis=0, ignore_index=True)
    load_concatenated = pd.concat(load_dataframes, axis=0, ignore_index=True)

    # Combine 'gen' and 'load' DataFrames
    combined_dataframe = pd.concat([gen_concatenated, load_concatenated], axis=0, ignore_index=True)

    return combined_dataframe

# Provide the path to your folder containing the CSV files
folder_path = 'data/raw_data/'
data = read_and_concatenate(folder_path)

# Optionally, save the combined DataFrame to a new CSV file
#combined_df.to_csv('combined_dataset.csv', index=False)
```

```python
def create_dataset(dataset, look_back=1):
    dataX, dataY = [], []
    for i in range(len(dataset)-look_back):
        a = dataset[i:(i+look_back), :]
        dataX.append(a)
        dataY.append(dataset[i + look_back, -1])  # Assuming the last column is the target variable
    return np.array(dataX), np.array(dataY)

# Assuming df is your DataFrame and DE_surplus is the target variable
look_back = 3  # Adjust as needed
dataset = de_df.values
X, y = create_dataset(dataset, look_back)

print(X.shape)
print(y.shape)

print(X[0])
print(y[0])
```
